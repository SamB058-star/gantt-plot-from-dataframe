```python
#importing libraries
import os, glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from datetime import date

#Names of columns:
colnames=['date', 'S1counts', 'S1SM', 'S1Perm', 'S1Temp', 'S1Volt',
        'S2counts', 'S2SM', 'S2Perm', 'S2Temp', 'S2Volt',
        'S3counts', 'S3SM', 'S3Perm', 'S3Temp', 'S3Volt',
        'S4counts', 'S4SM', 'S4Perm', 'S4Temp', 'S4Volt',
        'S5counts', 'S5SM', 'S5Perm', 'S5Temp', 'S5Volt',
        'S6counts', 'S6SM', 'S6Perm', 'S6Temp', 'S6Volt',
        'S7counts', 'S7SM', 'S7Perm', 'S7Temp', 'S7Volt',
        'S8counts', 'S8SM', 'S8Perm', 'S8Temp', 'S8Volt',
        'S9counts', 'S9SM', 'S9Perm', 'S9Temp', 'S9Volt',
        'S10counts', 'S10SM', 'S10Perm', 'S10Temp', 'S10Volt']  #D:\\Stations_data\\Bansathi\\Bansathi_all_months_copy

#matching the file directories
files_path = 'D:\\Stations_data\\Bansathi\\Bansathi_all_months_copy'
os.chdir(files_path)

#matching the file formats in that folder
extn = '.csv'
list_of_files = [i for i in glob.glob(f"*{extn}")]
#print(list_of_files)

mydateparser = lambda x: pd.datetime.strptime(x, "%d.%m.%Y %H:%M:%S")
df = pd.concat([pd.read_csv(f, sep = ',', skiprows = 1, skipfooter = 1, engine = 'python', date_parser = mydateparser,
                 header = None, names = colnames, skip_blank_lines = False) 
      for f in list_of_files], join = 'inner' )
#print(df.tail())
#df['date'] = df['date'].drop_duplicates(inplace = True)
#df.shape
#df.set_index('date')
```
